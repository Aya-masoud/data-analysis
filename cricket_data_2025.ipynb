```python
data_cat=[i for  i in data.columns if data[i].dtype =="O"]
data_cat.remove("Player_Name")
data_cat
```

```python
data["Highest_Score"]=data.Highest_Score.str.replace("*","",regex=False).astype(np.float32)
```

```python
data.head(20)
```

```python
data.Centuries=data.Centuries.astype(np.float16)
data.Centuries.describe()
```

```python
data["Not_Outs"]=data.Not_Outs.astype(np.float16)
data["Not_Outs"]=data.Not_Outs.astype(np.int16)
data.Not_Outs.describe()
```

```python
data.columns
```

```python
data["Runs_Scored"]=data.Runs_Scored.astype(np.float16)
data["Runs_Scored"]=data.Runs_Scored.astype(np.int16)
data.Not_Outs.describe()
```

```python
def De_types_converter(cat_col):
    for col in cat_col:
        
        data[col]=data[col].astype(np.float32)
        
        
```

```python
De_types_converter(cat_col=data_cat)
data.dtypes
```

```python
def converter_fraction(val):
    try:
        if '/' in val:
            num, den = val.split('/')
            return float(num) / float(den)
        else:
            return float(val)
    except:
        return None  

data.Best_Bowling_Match = data.Best_Bowling_Match.astype(str).apply(converter_fraction)
```

```python
import seaborn as sns
top_runs = data.groupby('Player_Name')['Runs_Scored'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_runs.values, y=top_runs.index, palette="viridis")
plt.title("Top 10 Run Scorers")
plt.xlabel("Total Runs Scored")
plt.ylabel("Player Name")
plt.show()
```

```python
min_matches = 10  
batting_avg_df = data[data['Matches_Batted'] >= min_matches].groupby('Player_Name')['Batting_Average'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=batting_avg_df.values, y=batting_avg_df.index, palette="coolwarm")
plt.title(f"Top Batting Averages (Min {min_matches} Matches Batted)")
plt.xlabel("Batting Average")
plt.ylabel("Player Name")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x="Runs_Scored", y="Batting_Strike_Rate", hue="Player_Name", legend=False, palette="tab10")
plt.title("Strike Rate vs. Runs Scored")
plt.xlabel("Runs Scored")
plt.ylabel("Batting Strike Rate")
plt.show()
```

```python
top_players = data.groupby('Player_Name')['Runs_Scored'].sum().sort_values(ascending=False).head(3).index

plt.figure(figsize=(10,6))
for player in top_players:
    player_df = data[data['Player_Name'] == player]
    yearly_runs = player_df.groupby('Year')['Runs_Scored'].sum().sort_index()
    plt.plot(yearly_runs.index, yearly_runs.values, marker='o', label=player)

plt.title("Runs Over the Years for Top 3 Run Scorers")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.show()
```
