```python
total_sales = data['Sales'].sum()
total_quantity = data['Quantity'].sum()
average_discount = data['Discount'].mean()
total_profit = data['Profit'].sum()
profit_margin = (data['Profit'].sum() / data['Sales'].sum()) * 100

print(f"Total Sales: {total_sales:.2f}")
print(f"Total Quantity Sold: {total_quantity}")
print(f"Average Discount: {average_discount:.2%}")
print(f"Total Profit: {total_profit:.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
quarterly_kpi = data.groupby('Order_Date_quarter').agg({
    'Sales': 'sum',
    'Quantity': 'sum',
    'Profit': 'sum'
}).reset_index()
```

```python
quarterly_kpi.plot(x='Order_Date_quarter', y=['Sales', 'Quantity', 'Profit'], kind='bar', figsize=(10,6))
plt.title('Total Sales, Quantity & Profit by Quarter')
plt.ylabel('Value')
plt.grid(axis='y')
plt.show()
```

```python

category_summary = data.groupby(['Category', 'Sub_Category']).agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

category_summary['Profit_Margin'] = category_summary['Profit'] / category_summary['Sales']

print(category_summary)
```
